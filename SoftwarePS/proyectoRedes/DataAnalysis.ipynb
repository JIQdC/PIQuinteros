```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
fname="Wed May 27 19:09:55 2020"
sec=np.loadtxt(fname,delimiter=",",skiprows=1,usecols=[0])
nsec=np.loadtxt(fname,delimiter=",",skiprows=1,usecols=[1])
data=np.loadtxt(fname,delimiter=",",skiprows=1,usecols=[4])
#rx_qstate=np.loadtxt(fname,delimiter=",",skiprows=1,usecols=[5])
#tx_qstate=np.loadtxt(fname,delimiter=",",skiprows=1,usecols=[6])
#cl_qstate=np.loadtxt(fname,delimiter=",",skiprows=1,usecols=[7])
sec=sec+nsec*1e-9
sec=sec-sec[0]
```

```python

print(np.size(sec))
samples=np.size(sec)-1
'''
plt.figure(1)
plt.plot(sec[:samples],rx_qstate[:samples],label="rx_q")
plt.plot(sec[:samples],tx_qstate[:samples],label="tx_q")
plt.plot(sec[:samples],cl_qstate[:samples],label="cl_q")
plt.xlim(0,samples/10000)
plt.legend()
'''
plt.figure(2)
plt.scatter(sec[:samples],data[:samples])
plt.xlim(0,samples/10000)
```

```python
difs=np.zeros_like(sec)
for i in range(1,np.size(sec)):
    difs[i-1]=sec[i]-sec[i-1]
difs[np.size(difs)-1]=difs[np.size(difs)-2]

plt.figure(4)
plt.plot(sec[:samples],difs[:samples])
    
```
