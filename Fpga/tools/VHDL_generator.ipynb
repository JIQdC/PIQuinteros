```python
register_name = "fifo_flags_r"
base = int("60",16)
```

```python
## read registers

for i in range(0,16):
    print("--fifo flags "+str(i+1)+"")
    print("when x\""+str(hex(base+i)[2:])+"\" =>")
    print("    axi_rdata <= "+register_name+"("+str(32*(i+1)-1)+" downto "+str(32*i)+");")
```

```python
#write registers

for i in range(0,16):
    print("--debug w2w1 "+str(i+1))
    print("when x\""+str(hex(base+i)[2:])+"\" =>")
    print("    for byte_index in 0 to (32/8-1) loop")
    print("        if ( S_AXI_WSTRB(byte_index) = '1' ) then")
    print("            "+register_name+"("+str(32*i)+"+byte_index*8+7 downto "+str(32*i)+"+byte_index*8) <= S_AXI_WDATA(byte_index*8+7 downto byte_index*8);")
    print("        end if;")
    print("    end loop;")
```
